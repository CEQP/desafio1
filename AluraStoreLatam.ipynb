### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

##tienda.head()
```

```python
tienda.columns = tienda.columns.str.replace(' ', '_')
tienda2.columns = tienda.columns.str.replace(' ', '_')
tienda3.columns = tienda.columns.str.replace(' ', '_')
tienda4.columns = tienda.columns.str.replace(' ', '_')
import matplotlib
import matplotlib.pyplot as plt
```

#1. Análisis de facturación



##Ingreso total por cada tienda

```python
ingresos_t1 = tienda.Precio.sum()
ingresos_t2 = tienda2.Precio.sum()
ingresos_t3 = tienda3.Precio.sum()
ingresos_t4 = tienda4.Precio.sum()
```

```python
ingresos_t1
```

# 2. Ventas por categoría

```python
ventasxcat_t1 = tienda.groupby("Categoría_del_Producto")[["Producto"]].count().sort_values("Producto", ascending=False )
ventasxcat_t2 = tienda2.groupby("Categoría_del_Producto")[["Producto"]].count().sort_values("Producto", ascending=False )
ventasxcat_t3 = tienda3.groupby("Categoría_del_Producto")[["Producto"]].count().sort_values("Producto", ascending=False )
ventasxcat_t4 = tienda4.groupby("Categoría_del_Producto")[["Producto"]].count().sort_values("Producto", ascending=False )
```

```python
ventasxcat_t1
```

# 3. Calificación promedio de la tienda


```python
valorizacion_t1 = tienda.Calificación.mean()
valorizacion_t2 = tienda2.Calificación.mean()
valorizacion_t3 = tienda3.Calificación.mean()
valorizacion_t4 = tienda4.Calificación.mean()
```

```python
valorizacion_t4
```

# 4. Productos más y menos vendidos

```python
ventas_tienda1 = tienda.Producto.value_counts()
ventas_tienda2 = tienda2.Producto.value_counts()
ventas_tienda3 = tienda3.Producto.value_counts()
ventas_tienda4 = tienda4.Producto.value_counts()
```

```python
ventas_tienda1
```

# 5. Envío promedio por tienda

```python
cenvio_tienda1 = tienda.Costo_de_envío.mean()
cenvio_tienda2 = tienda2.Costo_de_envío.mean()
cenvio_tienda3 = tienda3.Costo_de_envío.mean()
cenvio_tienda4 = tienda4.Costo_de_envío.mean()
```

```python
cenvio_tienda2
```

#6. Graficos

```python
ing_total_xtienda = pd.DataFrame({
    "Tienda": ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"],
    "Ventas": [round(ingresos_t1, 2),
               round(ingresos_t2, 2),
               round(ingresos_t3, 2),
               round(ingresos_t4, 2)],
    "Evaluacion":[round(valorizacion_t1, 2),
                  round(valorizacion_t2, 2),
                  round(valorizacion_t3, 2),
                  round(valorizacion_t4, 2)],
    "Costo Despacho X":[round(cenvio_tienda1, 2),
                        round(cenvio_tienda2, 2),
                        round(cenvio_tienda3, 2),
                        round(cenvio_tienda4, 2)]
}).set_index("Tienda")
```

```python
ing_total_xtienda = ing_total_xtienda.round(2)
pd.set_option('display.float_format', lambda x: f'{x:,.2f}')
```

```python
ing_total_xtienda
```

```python
ing_total_xtienda.plot(y="Ventas", kind="bar", figsize=(10,6), color="blue", xlabel="Tienda", ylabel="Ventas MMill")
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(10,6))

# Barras (Costo despacho)
ax1.bar(range(len(ing_total_xtienda)), ing_total_xtienda["Ventas"], color="blue", label="Ventas")
ax1.set_ylabel("Ventas MMill")
ax1.set_xlabel("Tienda")
ax1.set_xticks(range(len(ing_total_xtienda)))
ax1.set_xticklabels(ing_total_xtienda.index, rotation=45)

# Línea (Evaluacion) - segundo eje Y
ax2 = ax1.twinx()
ax2.plot(range(len(ing_total_xtienda)), ing_total_xtienda["Evaluacion"], color="red", marker="o", linewidth=2, label="Evaluacion")
ax2.set_ylabel("Evaluacion", color="red")

fig.legend(loc="upper right", bbox_to_anchor=(0.1, 0.9))
plt.show()
```

```python
fig, ax2 = plt.subplots(figsize=(10,6))

# Barras (Ventas)
ax2.bar(range(len(ing_total_xtienda)), ing_total_xtienda["Costo Despacho X"], color="Skyblue", label="Costo Desp. Prom.")
ax2.set_ylabel("Costo Desp.Prom")
ax2.set_xlabel("Tienda")
ax2.set_xticks(range(len(ing_total_xtienda)))
ax2.set_xticklabels(ing_total_xtienda.index, rotation=45)

# Línea (Evaluacion) - segundo eje Y
ax3 = ax2.twinx()
ax3.plot(range(len(ing_total_xtienda)), ing_total_xtienda["Evaluacion"], color="red", marker="o", linewidth=2, label="Evaluacion")
ax3.set_ylabel("Evaluacion", color="red")

fig.legend(loc="upper right", bbox_to_anchor=(0.1, 0.9))
plt.show()
```
